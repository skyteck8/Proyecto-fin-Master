```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Cargar el dataset
df = pd.read_csv('/content/mental_health_faq.csv')
df = df.drop(columns=['Question_ID'])
# Dividir en entrenamiento y validación
train_df, val_df = train_test_split(df, test_size=0.1)

# Convertir a un formato compatible con HuggingFace Datasets
from datasets import Dataset

train_dataset = Dataset.from_pandas(train_df)
val_dataset = Dataset.from_pandas(val_df)
```

Tokenización del dataset

```python
from transformers import BertTokenizer

# Cargar el tokenizador de BERT
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

def tokenize_function(examples):
    return tokenizer(examples['Questions'], examples['Answers'], truncation=True, padding='max_length', max_length=128)

train_dataset = train_dataset.map(tokenize_function, batched=True)
val_dataset = val_dataset.map(tokenize_function, batched=True)
```

```python
from transformers import BertForSequenceClassification

# Cargar el modelo de BERT preentrenado
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)
```

```python
import pandas as pd

# Cargar el dataset
df = pd.read_csv('/content/mental_health_faq.csv')

# Eliminar columnas innecesarias
df = df.drop(columns=['Question_ID'])

# Crear una columna de etiquetas
# Supongamos que las respuestas correctas están alineadas con las preguntas
df['labels'] = 1

# Crear ejemplos negativos (preguntas y respuestas que no corresponden)
# Podemos simplemente desordenar las respuestas para crear ejemplos negativos
negative_examples = df.copy()
negative_examples['Answers'] = negative_examples['Answers'].sample(frac=1).reset_index(drop=True)
negative_examples['labels'] = 0

# Combinar ejemplos positivos y negativos
combined_df = pd.concat([df, negative_examples]).reset_index(drop=True)

# Verificar el DataFrame combinado
print(combined_df.head())
```

```python
from sklearn.model_selection import train_test_split

# Dividir en entrenamiento y validación
train_df, val_df = train_test_split(combined_df, test_size=0.1)

# Convertir a un formato compatible con HuggingFace Datasets
from datasets import Dataset

train_dataset = Dataset.from_pandas(train_df)
val_dataset = Dataset.from_pandas(val_df)
```

```python
from transformers import BertTokenizer

# Cargar el tokenizador de BERT
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

def tokenize_function(examples):
    return tokenizer(examples['Questions'], examples['Answers'], truncation=True, padding='max_length', max_length=128)

train_dataset = train_dataset.map(tokenize_function, batched=True)
val_dataset = val_dataset.map(tokenize_function, batched=True)

# Asegurarte de que la columna 'labels' esté presente
train_dataset = train_dataset.rename_column('labels', 'labels')
val_dataset = val_dataset.rename_column('labels', 'labels')
```

```python
from transformers import BertForSequenceClassification, TrainingArguments, Trainer

# Cargar el modelo de BERT preentrenado
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)

# Configurar los argumentos de entrenamiento
training_args = TrainingArguments(
    output_dir='./results',
    evaluation_strategy='epoch',
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
)

# Crear el entrenador
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
)

# Entrenar el modelo
trainer.train()
```

```python
# Evaluar el modelo
trainer.evaluate()
```

```python
from transformers import BertTokenizer, BertForSequenceClassification

# Cargar el modelo y el tokenizador entrenados
model = BertForSequenceClassification.from_pretrained('./fine-tuned-bert')
tokenizer = BertTokenizer.from_pretrained('./fine-tuned-bert')
```

```python
def predict(question, answer):
    # Tokenizar la entrada
    inputs = tokenizer(question, answer, return_tensors='pt', truncation=True, padding='max_length', max_length=128)

    # Hacer la predicción
    outputs = model(**inputs)
    logits = outputs.logits
    prediction = logits.argmax(-1).item()

    return prediction

# Ejemplo de pregunta y respuesta
question = "What are the symptoms of anxiety?"
answer = "Symptoms of anxiety include excessive worrying, feeling agitated, restlessness, fatigue, difficulty concentrating, irritability, tense muscles, and trouble sleeping."

# Predecir si la respuesta es correcta
prediction = predict(question, answer)

if prediction == 1:
    print("La respuesta es correcta.")
else:
    print("La respuesta no es correcta.")
```

```python
def predict_best_answer(question, answer):
    # Tokenizar la entrada
    inputs = tokenizer(question, answer, return_tensors='pt', truncation=True, padding='max_length', max_length=128)

    # Hacer la predicción
    outputs = model(**inputs)
    logits = outputs.logits
    prediction = logits.argmax(-1).item()

    return prediction

# Función para encontrar la mejor respuesta
def get_best_answer(question):
    possible_answers = [
        "Symptoms of anxiety include excessive worrying, feeling agitated, restlessness, fatigue, difficulty concentrating, irritability, tense muscles, and trouble sleeping.",
        "Anxiety can cause sweating, trembling, and a sense of impending doom.",
        "To manage anxiety, try practicing mindfulness, deep breathing exercises, or seeking professional help.",
        "Physical exercise and a healthy diet can also help reduce symptoms of anxiety."
    ]

    best_answer = None
    best_prediction = -1
    for answer in possible_answers:
        prediction = predict_best_answer(question, answer)
        if prediction == 1:
            best_answer = answer
            break  # Salir del bucle cuando se encuentre una respuesta correcta
    return best_answer

# Bucle interactivo del chatbot
while True:
    question = input("You: ")

    if question.lower() in ['exit', 'quit', 'q']:
        print("Chatbot: Goodbye!")
        break

    best_answer = get_best_answer(question)

    if best_answer:
        print(f"Chatbot: {best_answer}")
    else:
        print("Chatbot: I'm not sure about the answer to that question. Can you ask something else?")
```
