```python
import transformers
import torch

model_id = "meta-llama/Meta-Llama-3-70B-Instruct"

pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device_map="auto",
)

messages = [
    {"role": "system", "content": "You are a pirate chatbot who always responds in pirate speak!"},
    {"role": "user", "content": "Who are you?"},
]

prompt = pipeline.tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
)

terminators = [
    pipeline.tokenizer.eos_token_id,
    pipeline.tokenizer.convert_tokens_to_ids("<|eot_id|>")
]

outputs = pipeline(
    prompt,
    max_new_tokens=256,
    eos_token_id=terminators,
    do_sample=True,
    temperature=0.6,
    top_p=0.9,
)
print(outputs[0]["generated_text"][len(prompt):])
```

```python
import transformers
import torch

# Define el modelo y crea el pipeline
model_id = "meta-llama/Meta-Llama-3-8B-Instruct"

pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device_map="auto",
)

# Inicializa la conversación
messages = [
    {"role": "system", "content": "You are a pirate chatbot who always responds in pirate speak!"},
]

while True:
    # Solicita la entrada del usuario
    user_input = input("You: ")

    # Verifica si el usuario desea salir
    if user_input.lower() == "exit":
        print("Conversation ended.")
        break

    # Añade el mensaje del usuario a la conversación
    messages.append({"role": "user", "content": user_input})

    # Genera el prompt para la conversación
    prompt = pipeline.tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )

    # Define los terminadores
    terminators = [
        pipeline.tokenizer.eos_token_id,
        pipeline.tokenizer.convert_tokens_to_ids("")
    ]

    # Genera la respuesta del modelo
    outputs = pipeline(
        prompt,
        max_new_tokens=256,
        eos_token_id=terminators,
        do_sample=True,
        temperature=0.6,
        top_p=0.9,
    )

    # Obtiene y muestra la respuesta del asistente
    assistant_response = outputs[0]["generated_text"][len(prompt):]
    print(f"Pirate Bot: {assistant_response.strip()}")

    # Añade la respuesta del asistente a la conversación
    messages.append({"role": "assistant", "content": assistant_response.strip()})
```
